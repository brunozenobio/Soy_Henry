# Ejercicio 104

Usar `groupby` para agrupar datos por una columna específica y contar la cantidad de elementos.

## Solución:

```python
df = pd.read_csv('./employees.csv')
df
```

```python

# El código agrupa el DataFrame `df` por la columna 'first_name', luego calcula el tamaño de cada grupo y finalmente ordena los grupos en orden descendente según su tamaño.

df.groupby('first_name').size().sort_values(ascending=False)
```

# Ejercicio 105

Usar la función `agg` para realizar varias operaciones por un eje de un `DataFrame`.

## Solución:

```python
# `df['salary'].agg(np.mean)` está calculando la media (promedio) de los valores en la columna 'salary' del DataFrame `df`.
df['salary'].agg(np.mean)
```

```python
# La función `df.agg(['max','min'])` se usa para calcular los valores máximo y mínimo para cada columna en un DataFrame `df`. Devuelve un nuevo DataFrame con los valores máximo y mínimo para cada columna.
df.agg(['max','min'])
```

```python
# El código `df.agg({'salary':['min','mean'],'employ_id':'min'})` está realizando operaciones de agregación en el DataFrame `df`.
df.agg({'salary':['min','mean'],'employ_id':'min'})
```

 # Ejercicio 140
 
 Crear un `DataFrame` a partir de arreglos `NumPy` y crear columnas e índices.
 
 ## Solución

```python
array = np.random.randint(5,2000,(8,5))
array
```

```python
columnas=['x','y','z','u','v']
indices = ['a','b','c','d','e','e','f','g']
```

```python
df_array = pd.DataFrame(data=array,columns=columnas,index=indices)
df_array
```

```python
print(f'Maximo por columna: \n{df_array.max()}')
print(f'Promedio por columna: \n{df_array.mean()}')
print(f'Valor medio por columna: \n{df_array.median()}')
print(f'Desviacion estandar por columna: \n{df_array.std()}')
```
