```python
import pandas as pd
import numpy as np
```

```python
nombre = ['Oliva', 'Daniela', 'Juan', 'Germán', 'Edward', 'Alex', 'Julio', 
          'Edgar', 'Angie', 'Irlesa']
puntaje = [11.5, 8, 15.5, np.nan, 8, 19, 13.5, np.nan, 8, 18]
intentos = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
califico = ['Sí', 'No', 'Sí', 'No', 'No', 'Sí', 'Sí', 'No', 'No', 'Sí']

indices = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

jugadores = {'nombre': nombre, 'puntaje': puntaje, 'intentos': intentos, 
             'califico': califico}

df = pd.DataFrame(data=jugadores, index=indices)

df
```

# Ejercicio 62

Crear un objeto `DataFrame` a partir de objetos `Series`.


```python
serie1 = pd.Series(['Mochila','Carpeta','Regla','Mesa','Celular'])
serie2 = pd.Series(['Negro','Rojo','Blanco','Marron','Azul'])
```

```python
precios = pd.Series(np.random.sample(5))
precios
```

```python

indices = ['a','b','c','d','e']
df_elem =pd.DataFrame( {
    'Elemento':serie1,
    'Color' : serie2,
    'Precio' : precios
})
df_elem.index=indices
df_elem
```

# Ejercicio 78

Ordenar en modo descendente un `DataFrame` por diferentes columnas.

## Solución:

```python
df.loc['g','puntaje'] = 17.0
df.loc['g','nombre'] = 'Juan'
```

```python
df.sort_values('nombre', ascending = False)
```

```python
df.sort_values(['nombre','puntaje'], ascending = False) # Al tambien ordenar por puntaje se invirtio el orden que antes
```
