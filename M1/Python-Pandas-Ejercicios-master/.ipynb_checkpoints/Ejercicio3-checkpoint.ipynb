# Ejercicio 103

Generar una representación de los elementos de un `DataFrame` con `applymap`.

## Solución:

```python
import pandas as pd 
import numpy as np
```

```python
numeros = [[3.1415, 2.717], [1.41, 1]]

df = pd.DataFrame(numeros)
df
```

```python
df.applymap(lambda z:z**2)
```

# Ejercicio 104

Usar `groupby` para agrupar datos por una columna específica y contar la cantidad de elementos.

## Solución:

```python
df = pd.read_csv('./employees.csv')
df
```

```python

df.groupby('first_name').size().sort_values(ascending=False)
```

# Ejercicio 105

Usar la función `agg` para realizar varias operaciones por un eje de un `DataFrame`.

## Solución:

```python
df['salary'].agg(np.mean)
```

```python
df.agg(['max','min'])
```

```python
df.agg({'salary':['min','mean'],'employ_id':'min'})
```

# Ejercicio 109

## Solución:

```python
(df != 100 ).all() # Todos los valores de la columna deben cumplir
```

```python
(df == 6400).any() # Si al menos un registro de alguna columna cumple la condicion devuelve true en la columna
```
