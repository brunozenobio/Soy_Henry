# Ejercicio 85

```python
df['puntaje'].where(df['puntaje']>200,0,inplace = True) # LOS QUE NO CUMPLAN LA CONDICION SE MODIFICAN
# EL METODO WHERE TAMBIEN PUEDE SER APLICADO A TODO EL DF  CAMBIANDO LOS VALORES DE TODAS LAS COLUMNAS CUANDO EL VALOR EN LA COLUMNA INDICADA CUMPLA LA CONDICION
# EJ df.where(df['puntaje']>200,0,inplace = True) PONE 0 EN TODAS LAS FILAS DONDE EL VALOR DE PUNTAJE SEA MENOR A 200
```

```python
df # COMO SE VE LA FUNCION ANTERIOR NO MODIFICA
# USAR INPLACE
```
