1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r',encoding="latin-1")
```

```python
columnas = archivo.readline().split('|')
```

```python
columnas
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
# GUARDO LOS VALORES EN UN DICCIONARIO
for line in archivo:
    line = line.strip()
    palabras = line.split("|")
    dicc_emisiones['cod_pais'].append(palabras[0])
    dicc_emisiones['nom_pais'].append(palabras[1])
    dicc_emisiones['region'].append(palabras[2])
    dicc_emisiones['anio'].append(palabras[3])
    dicc_emisiones['co2'].append(palabras[4])
    dicc_emisiones['co2_percapita'].append(palabras[5])
archivo.close()
```

```python
#SUMO LOS VALORES None
co2_none = 0
co2_perc_none = 0
nombre_none = 0
codigo_none = 0
anio_none = 0
region_none = 0
for ind,valor in enumerate(dicc_emisiones['co2']):
    if valor == None:
        co2_none += 1
for ind,valor in enumerate(dicc_emisiones['co2_percapita']):
    if valor == None:
        co2_perc_none += 1
for ind,valor in enumerate(dicc_emisiones['anio']):
    if valor == None:
        anio_none += 1
for ind,valor in enumerate(dicc_emisiones['cod_pais']):
    if valor == None:
        codigo_none += 1
for ind,valor in enumerate(dicc_emisiones['nom_pais']):
    if valor == None:
        nombre_none += 1
for ind,valor in enumerate(dicc_emisiones['region']):
    if valor == None:
        region_none += 1

print(f'Valores faltantes:\n Codigo Pais {codigo_none} \n Nombre Pais {nombre_none} \n Region {region_none} \n Año {anio_none}\n Co_2 valores faltantes:{co2_none} \n Co2_percapita valores faltantes:{co2_perc_none}')
```

```python
#SUMO LOS DATOS QUE CUMPLAN LA CONDICION
suma_co2_abw = 0
for ind,val in enumerate(dicc_emisiones['region']):
    if val == 'América Latina y Caribe' and dicc_emisiones['anio'][ind] == 2010 and dicc_emisiones['co2'][ind] != None:
        suma_co2_abw += dicc_emisiones['co2'][ind]
print(f'Las emisiones totales en 2010  en la region de America Latina y Caribe fueron de: {suma_co2_abw}')
```
