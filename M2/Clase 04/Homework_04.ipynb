```python
from math import factorial as fc
import math 
import matplotlib.pyplot as plt
```

```python
def dist_bin(n,k,p):
    return (fc(n) / (fc(k) * fc(n-k))) * (pow(p,k) *pow((1-p),n-k)) 
def poisson(mu,x):
    return pow(mu,x) * pow(math.e,-mu) / (fc(x))
```

En una encuesta realizada por Gallup Organization, se les preguntó a los interrogados, “Cuál es el deporte que prefieres ver”. Futbol y básquetbol ocuparon el primero y segundo lugar de preferencia (www.gallup.com, 3 de enero de 2004). Si en un grupo de 10 individuos, siete prefieren futbol y tres prefieren básquetbol. Se toma una muestra aleatoria de tres de estas persona

```python
from math import comb

def hiper_geo(N,n,r,x):
    return comb(r,x)*comb(N-r,n-x) / comb(N,n)
    
```

```python
# Siguiente la corriente frecuentista
# A partir de una muestra de 3 con que probabilidad dos prefieren futbol
# x = 2 numero de exitos
# n = 3
# p exito = p_futbol
N = 10
n = 3
r = 7 # Numero de exitos de la poblacion
x = 2 # Numero de exitos de la muestra


# Como se toma una poblacion de 10 con 7 exitos y 3 fracasos y luego una muestra de estos de 3 sera condicionada
# Por lo tanto se usara una distribucion hipergeometrica


print(f'La probabilidad que de un grupo de 3 dos prefieran futbol es: {hiper_geo(N,n,r,x)}')
```

```python

print(f'La probabilidad que de un grupo de 3 la mayoria prefiera futbol es: {hiper_geo(N,n,r,x) + hiper_geo(N,n,r,3)}')
```

La calificación promedio de los estudiantes de Henry es de 78 con una desviación estandar de 25. Cual es la probabilidad de tener estudiantes con calificaciones mayores o iguales 90.

```python
import math
import numpy as np
```

```python
def dist_normal(mean,desv,x):
    factor_raiz = math.sqrt(2*math.pi) 
    factor = 1 / (desv*factor_raiz)
    return factor * pow(math.e,-pow((x-mean) / desv,2)/2)
def dist_normal_estandar(mean,desv,x):
    z = (x-mean)/desv
    factor_raiz = math.sqrt(2*math.pi) 
    factor = 1 / (desv*factor_raiz)
    return factor * pow(math.e,-0.5 * (z**2))
```

```python
media = 78
desv = 25
# prob x>=90
# 1 - prob(entre 0 y 90)
x = np.linspace(1,200,2000)

# Estandarizo x

z = ((x-media) / desv)
y = dist_normal_estandar(media,desv,x)
```

```python
plt.plot(z,y)
plt.xlabel('Eje z distancia(z*sigma) de la media')
```
